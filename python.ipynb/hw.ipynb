```python
class room():
  def __init__(self,num,desc):
    self.room_num=num
    self.description=desc

  def __repr__(self):
    a="Room({},{})".format(self.room_num,self.description)
    return a

class thing():
  def __init__(self,name,self2):
    self.name=name
    self.location=self2

  def __str__(self):
    return "The "+self.name+" is in the "+self.location.description

class openable(thing):
  def __init__(self,name,self2,open=False):
    super().__init__(name,self2)
    self.openable=open

  def __str__(self):
    if self.openable==False:
      return "The "+self.name+" in "+self.location.description+" is closed"
    else:
      return "The "+self.name+" in "+self.location.description+" is open"

  def try_open(self):
    if self.openable==False:
      self.openable=True
      print("RETURNED TRUE")
      return True
    else:
      print("RETURNED FALSE")
      return False

class lockable(openable):
  def __init__(self,name,self1,self2,state=False,lck=False):
    super().__init__(name,self1,state)
    self.is_locked=lck
    self.key=self2

  def __str__(self):
    if self.openable==True:
      return "The "+self.name+" in "+self.location.description+" is open"
    elif self.openable==False and self.is_locked==False:
      return "The "+self.name+" in "+self.location.description+" is closed but unlocked"
    elif self.openable==False and self.is_locked==True:
      return "The "+self.name+" in "+self.location.description+" is locked"

  def try_open(self):
    if self.is_locked==True:
      print("RETURNED FALSE")
      return False
    elif self.is_locked==False:
      super().try_open()

  def try_unlock_with(self,self1):
    if(self.openable==True):
      print("RETURNED FALSE")
      return False
    elif self.key==self1:
      print("RETURNED TRUE")
      self.openable=True
      return True
    else:
      print("RETURNED FALSE")
      return False
```

```python
#TESTING
kitchen=room(12,"Kitchen")
master_key=thing('key',kitchen)
wrong_key=thing('rusted key',kitchen)
closet=openable('closet',kitchen)
door=openable('door',kitchen,True)
closet_lock=lockable('closet',kitchen,master_key)
cabinet_lock=lockable('cabinet',kitchen,master_key,False,True)
door=lockable('door',kitchen,master_key,True)
handle=lockable('handle',kitchen,master_key,False,True)
#PRINTING INITIAL STATUS
print(kitchen)
print(master_key)
print(wrong_key)
print(closet)
print(door)
print(closet_lock)
print(cabinet_lock)
print(door)
print()
#PRINTING STATUS AFTER CHECK
print("STATUS AFTER CHECK")
print()
closet.try_open()
print(closet)
door.try_open()
print(door)
closet_lock.try_open()
print(closet_lock)
cabinet_lock.try_open()
print(cabinet_lock)
door.try_open()
print(door)
print()
#PRINTING STATUS AFTER UNLOCK TEST
print("STATUS AFTER UNLOCK CHECK")
print()
closet_lock.try_unlock_with(master_key)
print(closet_lock)
cabinet_lock.try_unlock_with(wrong_key)
print(cabinet_lock)
cabinet_lock.try_unlock_with(master_key)
print(cabinet_lock)
handle.try_unlock_with(master_key)
print(handle)
```
