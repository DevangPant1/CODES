```python
t=np.random.randint([1,2,3,4,5])
y=np.random.randint([1,2,3,4,5])
plt.style.use('Solarize_Light2')
plt.plot(t,y,linestyle='dashed',marker='*',markersize=10)
plt.title("RANDOM")
plt.xlabel("RANDOMX")
plt.ylabel("RANDOMY")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
mobile=['samsung','apple','blackberry','xiaomi']
price=[30000,100000,35000,10000]
plt.figure(figsize=(10,5))
plt.bar(mobile,price,color='red',width=0.5)
plt.show()
```

```python
classes=['c++','c','java','python']
price=[10000,20000,30000,40000]
price_increase=[2000,3000,2500,5000]
plt.figure(figsize=(15,8))
plt.bar(classes,price,color='red')
plt.bar(classes,price_increase,bottom=price,color='black')
plt.show()
```

```python
fruit=['apple','banana','mango','kiwi']
number=np.array([50,19,45,60])
number_increase=np.array([3,6,9,12])
number_increase1=np.array([4,8,12,16])
plt.figure(figsize=(5,5))
plt.bar(fruit,number, color='green')
plt.bar(fruit,number_increase,bottom=number,color='red')
sum=number+number_increase
"""for i in range(len(number)):
  sum.append(number[i]+number_increase[i])"""
plt.bar(fruit,number_increase1,bottom=sum,color='blue')
plt.show()
```

#histogram chart


```python
import numpy as np
import matplotlib.pyplot as plt
age=np.random.normal(50,15,200)
plt.hist(age)
plt.show()
```

```python
fruit=np.array([50,23,67,90])
label1=['apple','banana','orange','mango']
plt.pie(fruit, labels=label1)
plt.show()
```

```python
activity=np.array([10,20,30,40)
label=['eating','walking','running','reading']
plt.pie(activity, labels=label)
plt.show()
```

```python
height=np.random.normal(150,50,100)
plt.hist(height)
plt.show()
```

```python
activity=np.array([10,20,30,40])
label=['eating','walking','running','reading']
plt.pie(activity, labels=label, startangle=180, explode=[0,0,0,0.1], colors=['#000000','blue','yellow','pink'])
plt.legend(title='activities',bbox_to_anchor=(1,1.1))
plt.show()
```
