```python
import requests
from bs4 import BeautifulSoup
url='https://www.flipkart.com/search?q=laptop&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off'
r=requests.get(url)
soup=BeautifulSoup(r.content,'html5lib')
laptop_names=soup.find_all('div',attrs={'class','_4rR01T'})
```

```python
lap=[]
for i in laptop_names:
  lap.append(i.text)
print(lap)
```

```python
price_not=soup.find_all('div',attrs={'class','_30jeq3 _1_WHN1'})
price=[]
for j in price_not:
  c=j.text.replace(',','')
  price.append(int(c[1:]))
print(price)
```

#FLIPKART


```python
import requests
url='https://www.flipkart.com/search?q=laptop&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off'
r=requests.get(url)
soup=BeautifulSoup(r.content,'html5lib')
names=soup.find_all('div',attrs={'class':'_4rR01T'})
ref_name=[i.text.split('-')[0] for i in names]
for i in ref_name:
  print(i)
```

```python
price=soup.find_all('div',attrs={'class','_30jeq3 _1_WHN1'})
final=[int(i.text.replace(',','').replace('₹','')) for i in price]
for i in final:
  print(i)
```

```python
import matplotlib.pyplot as plt
plt.plot(ref_name[:3],final[:3])
plt.title("PRICE AND NAME")
plt.xlabel("NAME")
plt.ylabel("PRICE")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
plt.figure(figsize=(100,100))
plt.bar(ref_name,final)
plt.show()
```

```python
plt.hist(price)
plt.show()
```
